```python
#1.ans) for loop is used in following situations:
    
   # a)when we want to iterate over a fixed sequence of values, such as a list, tuple, or range of numbers.
   # example
for i in range(10):
    print(i)
    #This will print the numbers 0 to 9.
    
   # b)when we know exactly how many times you want to iterate.
   # example
for i in range(3):
    print("hi")
   # This will print "hi" three times.
    
   # c)when we want to iterate over the elements of a collection, such as a list, tuple, or dictionary.
   # example
    fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
    #This will print each element of the fruits list.
    
   # While loop is used in following situation:
        
    #a) You want to repeat a block of code until a certain condition is met.
      #Example: 
x = 1
while x < 10:
    print(x)
    x += 1
    #This will print the numbers 1 to 9.
    
    #b) when we don't know how many times you need to iterate, and you want to keep looping until a certain condition is met.
    #example
while True:
    answer = input("Enter 'yes' or 'no': ")
    if answer == "yes":
        print("You entered 'yes'.")
        break
    elif answer == "no":
        print("You entered 'no'.")
        break
    else:
        print("Invalid input. Please try again.")
        #This will keep asking the user to enter "yes" or "no" until they do so.
        
        
      #c) when we want to iterate over a sequence of values, but you need to control the iteration manually.
       #Example: 
numbers = [1, 3, 5, 7, 9]
i = 0
while i < len(numbers):
    print(numbers[i])
    i += 1
        #This will print each element of the numbers list.
    
        

    
    
```

```python
#2 ans)
#  The sum and product of the first 10 natural numbers using for loop:

#Define variables for the sum and product
sum = 0
product = 1

# Loop through the first 10 natural numbers
for i in range(1, 11):
    sum += i
    product *= i

# Print the sum and product
print("The sum of the first 10 natural numbers is:", sum)
print("The product of the first 10 natural numbers is:", product)

# The sum and product of the first 10 natural numbers using a while loop

# Define variables for the sum and product
sum = 0
product = 1

i = 1

# Loop through the first 10 natural numbers using a while loop
while i <= 10:
    sum += i
    product *= i
    i += 1

# Print the sum and product
print("The sum of the first 10 natural numbers is:", sum)
print("The product of the first 10 natural numbers is:", product)

```

```python
#3 ans) program to calculate electricity bill

unit = int(input("how many units of electricity did u  consume "))
if unit > 0 and unit < 100:
    bill =unit*4.5
    print("bills=",bill)
elif unit>100 and unit<200 :
    bill=(100*4.5)+(unit-100)*6
    print("bills=",bill)
elif unit>200 and unit<300:
    bill=(100*4.5)+(100*6)+(unit-200)*10
    print("bill=",bill)
elif unit>300:
    bill=(100*4.5)+(100*6)+(100*10)+(unit-300)*20
    print("bill=",bill)
    
```

```python
#4 ans) The cubes of numbers 1 to 100 that are divisible by 4 or 5 using for loop
cubes_list = []
for i in range(1, 101):
    cube = i ** 3
    if cube % 4 == 0 or cube % 5 == 0:
        cubes_list.append(cube)

# Print the list of cubes
print(cubes_list)

#  The cubes of numbers 1 to 100 that are divisible by 4 or 5 using while loop
cubes_list = []
i = 1
while i <= 100:
    cube = i ** 3
    if cube % 4 == 0 or cube % 5 == 0:
        cubes_list.append(cube)
    i += 1

# Print the list of cubes
print(cubes_list)

```
